Sua miss√£o √© criar um c√≥digo que, ao receber a data e hora de um compromisso em um fuso hor√°rio (GMT), seja capaz de converter essa data e hora para o fuso hor√°rio local de outros tr√™s pa√≠ses: Brasil (GMT-3), Jap√£o (GMT+9) e Estados Unidos (Nova York, GMT-4).

Receba a data e hora de um compromisso (em GMT) e converta para os fusos hor√°rios locais do Brasil, Jap√£o e Nova York.
Exiba uma mensagem para o usu√°rio indicando os hor√°rios em cada pa√≠s, facilitando a organiza√ß√£o da agenda.

Sa√≠da Esperada:

"Seu compromisso ser√° √†s 12:00 no Brasil, √†s 00:00 no Jap√£o (do dia seguinte), e √†s 11:00 em Nova York."

```python
compromisso_gmt = '2024-10-10 15:00:00'
```

```python
from datetime import datetime
import pytz

def converter_compromisso(compromisso_gmt_str):
   gmt = pytz.timezone('GMT')
   compromisso_gmt = datetime.strptime(compromisso_gmt_str, '%Y-%m-%d %H:%M:%S')
   compromisso_gmt = gmt.localize(compromisso_gmt)

   brasil = pytz.timezone('America/Sao_Paulo')
   japao = pytz.timezone('Asia/Tokyo')
   nova_york = pytz.timezone('America/New_York')

   compromisso_brasil = compromisso_gmt.astimezone(brasil)
   compromisso_japao = compromisso_gmt.astimezone(japao)
   compromisso_nova_york = compromisso_gmt.astimezone(nova_york)

   horario_brasil = compromisso_brasil.strftime('%H:%M')
   horario_japao = compromisso_japao.strftime('%H:%M')
   horario_nova_york = compromisso_nova_york.strftime('%H:%M')

   print(f"Seu compromisso ser√° √†s {horario_brasil} no Brasil, √†s {horario_japao} no Jap√£o (do dia seguinte), e √†s {horario_nova_york} em Nova York.")

compromisso_gmt = '2024-10-10 15:00:00'
converter_compromisso(compromisso_gmt)
```
