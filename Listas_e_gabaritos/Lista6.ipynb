### **Quest√£o 3**

b) Verifique as lojas que tiveram lucro abaixo da m√©dia e faturamento menor que 200 mil reais.

```python
abaixo_media_menos200 = df[(df['lucro'] < lucro_medio) & (df['faturamento'] < 200000)]['cod_loja'].tolist()
abaixo_media_menos200
```

c) Mostre as lojas das cidades de Curitiba e Bras√≠lia.

```python
lojas_cur_bras = df[df['cidade'].isin(['Curitiba', 'Bras√≠lia'])]
lojas_cur_bras
```

d) Mostre as linhas das lojas com c√≥digo entre LJ-250 e LJ-260 (inclusive).

```python
df['codigo_numerico'] = df['cod_loja'].str.extract(r'(\d+)$').astype(int)
lojas_250_260 = df[(df['codigo_numerico'] >= 250) & (df['codigo_numerico'] <= 260)]

lojas_250_260 = lojas_250_260.drop(columns=['codigo_numerico'])

lojas_250_260
```
